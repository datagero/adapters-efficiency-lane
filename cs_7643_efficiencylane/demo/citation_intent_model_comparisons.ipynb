```python
from transformers import RobertaTokenizer, RobertaConfig
from transformers import RobertaForSequenceClassification, Trainer, EvalPrediction, TrainingArguments
from sklearn.metrics import accuracy_score, f1_score
from adapters import AutoAdapterModel, RobertaAdapterModel, AdapterTrainer
from gp import setup_dataset
import torch
import numpy as np
import matplotlib.pyplot as plt

dataset, num_labels = setup_dataset('citation_intent')

model_name = 'roberta-base'
dataset_name = 'citation_intent'
model_type = 'base'

# Roberta Base model counts

fine_tuning_model = RobertaForSequenceClassification.from_pretrained(
        model_name, num_labels=num_labels,
        )

# Iterate through the model's layers to get the parameter count for each layer
layer_info = []

for name, param in fine_tuning_model.named_parameters():
    # Get the number of parameters in each layer
    param_count = param.numel()
    # Collect layer name and parameter count
    layer_info.append((name, param_count))

# Print the total number of parameters
total_params = sum(param_count for _, param_count in layer_info)

print("Total Parameter Count:", total_params)

# Print the layer information with parameter counts
print("\nLayer-wise Parameter Counts:")
for name, param_count in layer_info:
    print(f"{name}: {param_count}")
```

```python
#Parameter count for Pfieffer adapter
adapter_name = model_type+"_"+dataset_name
config = RobertaConfig.from_pretrained(
        model_name,
        num_labels=num_labels
)
adapter_model = RobertaAdapterModel.from_pretrained(model_name,config=config)
adapter_model.add_adapter(adapter_name, config="seq_bn")
adapter_model.add_classification_head(
        adapter_name,
        num_labels=num_labels,
        overwrite_ok=True
)

# Iterate through the model's layers to get the parameter count for each layer
layer_info = []

for name, param in adapter_model.named_parameters():
    # Get the number of parameters in each layer
    param_count = param.numel()
    # Collect layer name and parameter count
    layer_info.append((name, param_count))

# Print the total number of parameters
total_params = sum(param_count for _, param_count in layer_info)

print("Total Parameter Count:", total_params)

# Print the layer information with parameter counts
print("\nLayer-wise Parameter Counts:")
for name, param_count in layer_info:
    print(f"{name}: {param_count}")
```
